```python
import urllib.request
from bs4 import BeautifulSoup

url = "http://www.alltime-athletics.com/m_10kok.htm"

with urllib.request.urlopen(url) as resp:
    processed_page = BeautifulSoup(resp.read(), "html.parser")

pre_tag = processed_page.find("pre")
values_text = pre_tag.text

# Split the text into individual lines
lines = values_text.split("\n")

values_array = []

for line in lines:
    # Remove any leading/trailing whitespaces
    line = line.strip()
    if line:
        # Split the line into values based on the pattern you provided
        values = line.split()


        name_parts = values[2:values.index(values[4])]
        name = ' '.join(name_parts)
        values = values[:2] + [name] + values[4:]


        if len(values[3]) != 3:
            # Concatenate the 3rd index with the 2nd index
            values[2] = values[2] + ' ' + values[3]
            del values[3]

        # Delete the value in the 6th index
        if len(values) > 5:
            del values[5]
        values_array.append(values)

for row in values_array:
    if len(row) > 6:
        joined_values = ' '.join(row[5:-1])
        row[5] = joined_values
        del row[6:-1]

for row in values_array:
    print(row)
```

```python
import csv

filename = 'men_10000m.csv'
with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = [
        'Place',
        'Time',
        'Athlete',
        'Country',
        'Date of Birth',
        'City',
        'Event Date'
    ]
    writer = csv.writer(csvfile)
    
    writer.writerow(fieldnames)  # Write the header row
    
    for row in values_array:       
        writer.writerow(row)
    
print(f"Data saved to {filename} successfully.")
```
